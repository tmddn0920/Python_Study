```python
# 8

import pandas as pd
college = pd.read_csv('College.csv')
college
```

```python
college2 = pd.read_csv("College.csv", index_col=0)
college3 = college.rename({"Unnamed: 0": "College"}, axis=1)
college3 = college3.set_index("College")
college = college3
college
```

```python
college.describe()
```

```python
import matplotlib.pyplot as plt
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]], figsize=(10,10))
plt.show()
```

```python
college.boxplot(column="Outstate", by="Private", figsize=(8, 6))
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'],[0,50,100],labels =['No','Yes'])
```

```python
college['Elite'].value_counts()
```

```python
college.boxplot(column="Outstate",by="Elite",figsize=(8, 6))
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
college['Apps'].hist(bins=10, ax=axes[0, 0])
college['Accept'].hist(bins=10, ax=axes[0, 1])
college['Enroll'].hist(bins=20, ax=axes[1, 0])
college['Outstate'].hist(bins=20, ax=axes[1, 1])
plt.tight_layout()
plt.show()
```
